```python
# 首先，确保'CO con: (ppm)'和'Ethylene con: (ppm)'列存在
if 'CO con: (ppm)' in data.columns and 'Ethylene con: (ppm' in data.columns:
    # 创建新列，使用条件表达式填充
    data['CO'] = data['CO con: (ppm)'].apply(lambda x: 0 if pd.isna(x) or x == 0 else 1)
    data['Ethylene'] = data['Ethylene con: (ppm'].apply(lambda x: 0 if pd.isna(x) or x == 0 else 1)
else:
    print("列名不存在")

# 现在df中包含了新列，原始数据未受影响
```

```python
import matplotlib.pyplot as plt

# 计算每个类别的数量
co_zero_count = (data['CO'] == 0).sum()
co_one_count = (data['CO'] == 1).sum()

ethylene_zero_count = (data['Ethylene'] == 0).sum()
ethylene_one_count = (data['Ethylene'] == 1).sum()

# 创建饼图
plt.figure(figsize=(10, 7))
plt.subplot(1, 2, 1)  # 1行2列的第一个图
plt.pie([co_zero_count, co_one_count], labels=['0 ppm', '>0 ppm'], autopct='%1.1f%%', startangle=90)
plt.title('CO')

plt.subplot(1, 2, 2)  # 1行2列的第二个图
plt.pie([ethylene_zero_count, ethylene_one_count], labels=['0 ppm', '>0 ppm'], autopct='%1.1f%%', startangle=90)
plt.title('Ethylene')

# 显示图表
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 假设df是你的DataFrame

# 定义一个函数，根据CO和Ethylene的值返回相应的数字
def combine_values(row):
    if row['CO'] == 0 and row['Ethylene'] == 0:
        return 0
    elif row['CO'] == 1 and row['Ethylene'] == 0:
        return 1
    elif row['CO'] == 0 and row['Ethylene'] == 1:
        return 2
    else:  # 这意味着两者都为1
        return 3

# 应用这个函数到DataFrame的每一行，并创建新列
data['target'] = data.apply(combine_values, axis=1)
```

```python
import matplotlib.pyplot as plt

# 计算每个combined_value的数量
value_counts = data['target'].value_counts()

# 创建柱状图
plt.figure(figsize=(8, 4))
plt.bar(value_counts.index, value_counts.values, color=['blue', 'green', 'red', 'purple'])

# 添加一些图表元素
plt.xlabel('Combined Values')
plt.ylabel('Counts')
plt.title('Distribution of Combined Values')
plt.xticks(value_counts.index)  # 确保x轴标签正确显示

# 显示图表
plt.tight_layout()
plt.show()
```
